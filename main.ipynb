```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
#  creating new column
labels = ['Young Adult', 'Adult','Middle-aged', 'Senior']
# df['age_group'] = pd.qcut(df['age'], bins=[0, 25, 40, 60, 100], labels=labels)
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df.head()
```

```python
df[['age', 'age_group']].head(10)
```

```python
# new column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Anually': 365,
    'Every 3 months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df = df.drop(columns=['promo_code_used'], axis=1)
# axis = 1 is used to specify that we want to drop a column, not a row. By default, drop() assumes you want to drop rows (axis=0), so we need to set axis=1 to indicate that we're dropping a column.
```
